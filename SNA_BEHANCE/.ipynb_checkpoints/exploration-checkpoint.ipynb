```python
import pandas as pd
import numpy as np
import gzip
import json
import itertools
import networkx as nx
import matplotlib.pyplot as plt

%matplotlib inline
#GOAL: ITEM RELATIONSHIP
#HCS Clustering algorithm
```

## Donwloading the Behance files, and converting them into a list. 
### We have two lists, One for appreciation items, with user, item and timestamp. The second one is for item's and owners.

```python
appreciation_path = 'data/Behance_appreciate_1M.gz'
appreciation_file = gzip.open(appreciation_path,'rb')
appreaciation_content=appreciation_file.read().decode('utf8')

owners_path = 'data/Behance_Item_to_Owners.gz'
owners_file = gzip.open(owners_path,'rb')
owners_content=owners_file.read().decode('utf8')

a_triplets = appreaciation_content.split('\n')
o_triplets = owners_content.split('\n')

appreciation_dict = []
owners_dict = []
for triple in a_triplets:
    v = triple.split()
    if len(v)==3:
        d = {
            'user_id':v[0],
            'item_id':v[1],
            'timestamp':v[2]
        }
        appreciation_dict.append(d)
limit = 10000
for triple in o_triplets[:limit]:
    v = triple.split()
    if len(v)==2:
        d = {
            'item_id':v[0],
            'user_id':v[1]
        }
        owners_dict.append(d)
```

```python
limit
```

```python
df_app = pd.DataFrame(appreciation_dict)
df_own = pd.DataFrame(owners_dict)
```

```python
owners_to_items = dict()
for tup in owners_dict:
    user_id = tup.get('user_id')
    item_id = tup.get('item_id')
    if user_id in owners_to_items:
        owners_to_items.get(user_id).append(item_id)
    else:
        owners_to_items[user_id] = [item_id]
```

```python
def combinations(item_list, combination_list=[]):

    if len(item_list)>=2:
        for i in range(1,len(item_list)):
            d = {
                    'source':item_list[0],
                    'target':item_list[i]
                }
            combination_list.append(d)
        item_list.pop(0)
        return combinations(item_list, combination_list)
    else:
        return item_list, combination_list

nodes = [i.get('item_id') for i in owners_dict]
edgelist = dict()
edge_s =[]
edge_t =[]
for i, item_list in enumerate(owners_to_items.values()):
    if len(item_list) >= 2:
        l, c = combinations(item_list)
        for d in c:
            edge_s.append(d.get('source'))
            edge_t.append(d.get('target'))
        
#edges = list(itertools.chain.from_iterable(edgelist))
```

```python
df_edges = pd.DataFrame(data=list(zip(edge_s,edge_t)),columns=['source','target'])
```

```python
nodes = list(set(df_edges['source'].unique().tolist() + df_edges['target'].unique().tolist()))
```

```python
edges = list(df_edges.itertuples(index=False, name=None))
```

```python
G = nx.Graph()
```

```python
G.add_nodes_from(nodes)
```

```python
G.add_edges_from(edges)
```

```python
options = {
    "node_size": 10,
    "edge_color": "grey",
    "linewidths": 1,
    "width": 1,
}
plt.figure(figsize=(22, 22))
nx.draw(G)
plt.show()
```

```python
sorted(dict(G.degree()).items(), key=lambda x: x[1])[-1]
```

```python
remove = [node for node,degree in dict(G.degree()).items() if degree < 20]
G.remove_nodes_from(remove)
```

```python
highest_dg_node = sorted(dict(G.degree()).items(), key=lambda x: x[1])[-1][0]
```

```python
chosen_nodes = list(G.neighbors(highest_dg_node))
```

```python
sub = nx.subgraph(G, chosen_nodes)
```
